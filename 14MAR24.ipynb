#### matplotlib
crea visualizaciones estáticas, animadas e interactivas en python
figure es lo que crea y puede tener uno o más axis.

- Para crear una figura, se utiliza el método subplot
- Componentes de una figura:
1. La figura
2. Lineas
3. Formas
4. Leyendas: si tengo más de una línea, me dice a qué corresponde
5. Axis
6. Cortes para ver solo una parte de interés
- Los ejes tiene unos métodos asociados: set_title(): titulo, set_xlabel(), set_ylabel(): nombre del eje x y y 



```python
import matplotlib.pyplot as plt #metodo que nos va a permitir graficar
import numpy as np

x=np.linspace(0,2*np.pi,200) #va de 0 a 2*pi con 200 puntos
print(x.size)
y=np.sin(x) #se le aplica el seno a cada una de las posiciones de x
print (y.size)

fig, ax=plt.subplots() #siempre tiene asociada la fig que es un cuadrado que tiene la información, ax son los ejes (AXIS)
ax.plot(x,y, label="Grafica seno") #lo que voy a graficas: asignación de los axis #label para leyendas
ax.set_xlabel("Eje x")
ax.set_ylabel("Eje y")
ax.set_title("Gráfica de la función seno(x)")
ax.legend() #tiene que estar para que se agregue el label al gráfico
plt.show() #mostrar la figura
```

```python
np.random.seed(19680801)
data={'a': np.arange(50),
      'c': np.random.randint(0,50,50),
      'd': np.random.randn(50)}

data['b']=data['a'] + 10 * np.random.randn(50)
data['d']= np.abs(data['d'])*100

fig,ax=plt.subplots(figsize=(5,2.7), layout='constrained')
ax.scatter('a','b',c='c',s='d', data=data)
ax.set_xlabel('entry a')
ax.set_ylabel('entry b')
```

- implicita,explicita(POO)

```python
def my_plotter(ax,data1,data2,param_dict):
    """
    A helper function to make a graph
    """
    out=ax.plot(data1,data2,**param_dict)
    return out


data1, data2, data3, data4=np.random.rand(4,100)
fig, (ax1,ax2)=plt.subplots(1,2,figsize=(5,2.7))

my_plotter(ax1,data1,data2,{'marker':'x'})
my_plotter(ax2,data3,data4,{'marker':'o'})
```

```python
x1=np.linspace(0,3*np.pi,200) #va de 0 a 2*pi con 200 puntos
y1=np.sin(x1) #se le aplica el seno a cada una de las posiciones de x
x2=np.linspace(0,2*np.pi,200)
y2=np.cos(x2)


fig, (ax1,ax2)=plt.subplots(2,1, figsize=(5,5)) #siempre tiene asociada la fig que es un cuadrado que tiene la información, ax son los ejes (AXIS) #DOS FILAS Y UNA SOLA COLUMNA
#figsize cambia el tamaño de la figura
ax1.plot(x1,y1, label="Grafica seno") #lo que voy a graficas: asignación de los axis #label para leyendas
ax2.plot(x2,y2, label="Grafica coseno")
ax1.set_xlabel("Eje x")
ax1.set_ylabel("Eje y")
ax1.set_title("grafica de la funcion sen")
ax1.legend()
ax2.set_xlabel("Eje x")
ax2.set_ylabel("Eje y")
ax2.set_title("grafica de la funcion cos")
ax2.legend()

plt.show() #mostrar la figura
```

```python
x1=np.linspace(0,3*np.pi,200) #va de 0 a 2*pi con 200 puntos
y1=np.sin(x1) #se le aplica el seno a cada una de las posiciones de x
x2=np.linspace(0,2*np.pi,200)
y2=np.cos(x2)


fig, (ax1,ax2)=plt.subplots(2,1, figsize=(5,5)) #siempre tiene asociada la fig que es un cuadrado que tiene la información, ax son los ejes (AXIS) #DOS FILAS Y UNA SOLA COLUMNA
#figsize cambia el tamaño de la figura
ax1.plot(x1,y1,'x' ,label="Grafica seno",color="pink",linewidth=3) #lo que voy a graficas: asignación de los axis #label para leyendas
ax2.plot(x2,y2, label="Grafica coseno",color="purple",linewidth=5,linestyle=":")
ax1.set_xlabel("Eje x")
ax1.set_ylabel("Eje y")
ax1.set_title("grafica de la funcion sen")
ax1.legend()
ax2.set_xlabel("Eje x")
ax2.set_ylabel("Eje y")
ax2.set_title("grafica de la funcion cos")
ax2.legend()

plt.show() #mostrar la figura
```

AGREGAMOS UN TEXTO, Y EL GRID

```python
x1=np.linspace(0,3*np.pi,200) #va de 0 a 2*pi con 200 puntos
y1=np.sin(x1) #se le aplica el seno a cada una de las posiciones de x
x2=np.linspace(0,2*np.pi,200)
y2=np.cos(x2)


fig, (ax1,ax2)=plt.subplots(2,1, figsize=(5,5)) #siempre tiene asociada la fig que es un cuadrado que tiene la información, ax son los ejes (AXIS) #DOS FILAS Y UNA SOLA COLUMNA
#figsize cambia el tamaño de la figura
ax1.plot(x1,y1,'x' ,label="Grafica seno",color="pink",linewidth=3) #lo que voy a graficas: asignación de los axis #label para leyendas
ax2.plot(x2,y2, label="Grafica coseno",color="purple",linewidth=5,linestyle=":")
ax1.set_xlabel("Eje x")
ax1.set_ylabel("Eje y")
ax1.set_title("grafica de la funcion sen")
ax1.legend()
ax1.text(4,0.50, "HOLA") #ACA SE AGREGA EL TEXTO, EN LA POSICION
ax1.grid()
ax2.set_xlabel("Eje x")
ax2.set_ylabel("Eje y")
ax2.set_title("grafica de la funcion cos")
ax2.legend()
ax2.grid()


plt.show() #mostrar la figura
```
