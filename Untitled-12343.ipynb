```python
import os
import pydicom
import matplotlib.pyplot as plt

# Ruta de la carpeta que contiene las imágenes DICOM
folder_path = r"C:/Users/griss/Downloads/archivosDCM"

# Obtener la lista de archivos DICOM en la carpeta
files = os.listdir(folder_path)
# Filtrar solo los archivos DICOM
dicom_files = [f for f in files if f.endswith(".dcm")]
# Ordenar los archivos por nombre para asegurar el orden correcto
dicom_files.sort()

# Lista para almacenar las imágenes DICOM
dicom_images = []

# Iterar sobre cada archivo DICOM, cargarlo y agregarlo a la lista
for file in dicom_files:
    # Ruta completa del archivo DICOM
    file_path = os.path.join(folder_path, file)
    # Cargar el archivo DICOM
    dicom_image = pydicom.dcmread(file_path)
    # Agregar la imagen DICOM a la lista
    dicom_images.append(dicom_image)

# Índice de la imagen actual
current_index = 0

# Función para mostrar la imagen actual
def show_image():
    global current_index
    plt.imshow(dicom_images[current_index].pixel_array, cmap=plt.cm.gray)
    plt.title(f"Imagen DICOM {current_index + 1}")
    plt.show()

# Función para avanzar a la siguiente imagen
def next_image(event):
    global current_index
    if event.key == 'right':
        current_index = (current_index + 1) % len(dicom_images)
        show_image()

# Mostrar la primera imagen
show_image()

# Conectar la función next_image al evento de presionar una tecla
plt.connect('key_press_event', next_image)
plt.show()

```

```python
import cv2
import numpy as np

# 1. Leer la imagen
image = cv2.imread("C:/Users/griss/Documents/Parcial 1 Info2/Imagen célula.jpg")

# 2. Convertir la imagen a escala de grises
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# 3. Aplicar un filtro de suavizado para reducir el ruido
blurred_image = cv2.GaussianBlur(gray_image, (5, 5), 0)

# 4. Aplicar umbralización para convertir la imagen en una imagen binaria
_, binary_image = cv2.threshold(blurred_image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# 5. Aplicar operaciones morfológicas para eliminar el ruido y conectar las regiones de interés
kernel = np.ones((5, 5), np.uint8)
morph_image = cv2.morphologyEx(binary_image, cv2.MORPH_CLOSE, kernel)

# 6. Encontrar y contar los contornos de las regiones de interés
contours, _ = cv2.findContours(morph_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
cell_count = len(contours)
print("Número de células encontradas:", cell_count)

# 7. Dibujar los contornos en la imagen original
contour_image = np.copy(image)
cv2.drawContours(contour_image, contours, -1, (0, 255, 0), 2)

# 8. Mostrar la imagen con los contornos resaltados y el número de células contadas
cv2.putText(contour_image, f"Células: {cell_count}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 2)
cv2.imshow("Conteo Celular", contour_image)
cv2.waitKey(0)
cv2.destroyAllWindows()

```
