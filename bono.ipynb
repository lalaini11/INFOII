```python
# importa fft y la inversa
from scipy.fftpack import fft  
#Importing numpy  
import numpy as np  
#create an array with random n numbers  
x = np.random.rand(20)
#Applying the fft function  
y = fft(x)  

print(x)
print(y) 
```

```python
import matplotlib.pyplot as plt #metodo que nos va a permitir graficar
import numpy as np

x=np.random.rand(20) 
y1=np.random.rand(20)
y2=fft(x) #transforma X


fig, (ax,ax2)=plt.subplots(1,2, figsize=(5,5)) #siempre tiene asociada la fig que es un cuadrado que tiene la información, ax son los ejes (AXIS) 
#figsize cambia el tamaño de la figura
ax.plot(x,y1,label="Señal sin transformar",color="pink",linewidth=5) #lo que voy a graficas: asignación de los axis #label para leyendas
ax2.plot(x,y2, label="Señal transformada",color="blue",linewidth=3,linestyle=":")
ax.set_xlabel("Tiempo") #Leyenda en x
ax.set_ylabel("Amplitud") #Leyenda en y
ax.set_title("Señal sin transformar") #Titulo de la grafica
ax.legend()
ax2.set_xlabel("Frecuencia") #Leyenda en x
ax2.set_ylabel("Amplitud") #Leyenda en y
ax2.set_title("Señal transformada") #Titulo de la grafica
ax2.legend()

plt.show() #mostrar la figura
```
